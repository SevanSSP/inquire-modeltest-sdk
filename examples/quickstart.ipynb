# Quickstart
Examples to get you up and running in no time.

```python
# imports
from modeltestSDK import Client

# configurations
%matplotlib notebook

# initiate client with defaults
client = Client()
```

```python
# list available campaigns
campaigns = client.campaign.get()
```

```python
# have a look at the data (using Pandas here)
campaigns.to_pandas()
```

```python
# which tests were carried out for White Pearl? (it is the first campaign in the list above)
campaign = campaigns[0]
tests = campaign.tests()
tests.to_pandas().head() # just list the first ones
# you can also do it as `campaigns[0].tests().to_pandas()`
```

```python
# and the sensor setup?
sensors = campaign.sensors()
sensors.to_pandas().head(n=20) # just list the first 20.
```

```python
# let's have a closer look at the surge decay tests (the third in the above list)
# and the surge motion ("XPOS_WL")
test = tests[3]
test.timeseries(sensor_id="d9a51897-3511-462a-bb92-d2cc2b368e1d").plot()
```
